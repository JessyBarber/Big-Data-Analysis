```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as img
from sklearn.cluster import MiniBatchKMeans
```

```python
%matplotlib inline
goldCoast = img.imread("goldcoast.jpeg")
ax = plt.axes(xticks=[], yticks=[])
print(ax)
plt.imshow(goldCoast)
```

```python
goldCoast.shape
```

```python
#Contains red/blue/green which is ints 0-255
data = goldCoast / 255.0
#Reshape into 1D array
data = data.reshape(183 * 275, 3)
data.shape
```

```python
#K = 4
import warnings; warnings.simplefilter('ignore')
kmeans = MiniBatchKMeans(4)
kmeans.fit(data)
new_colors = kmeans.cluster_centers_[kmeans.predict(data)]
```

```python
goldCoast_recolored = new_colors.reshape(goldCoast.shape)
fig, ax = plt.subplots(1, 2, figsize=(16, 6), subplot_kw=dict(xticks=[],yticks=[]))
fig.subplots_adjust(wspace=0.05)
ax[0].imshow(goldCoast)
ax[0].set_title('Original Image', size=16)
ax[1].imshow(goldCoast_recolored)
ax[1].set_title('4-color Image', size=16)
```
